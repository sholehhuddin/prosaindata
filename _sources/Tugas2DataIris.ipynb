##PCA

setelah mendapatkan data utuh dari masing masing database  lakukan pca untuk mengurangi dimensi data agar perhitungan jadi lebih mudah

```python
import orchest
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings("ignore")
data = orchest.get_inputs()

df = data['data']

X = df.iloc[:, :-1]
y = df.iloc[:, 4]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
pca = PCA(n_components = 1)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
hasil_variance = pca.explained_variance_ratio_ # Proses pengecekan variance
type(X_train)

orchest.output((X_train,X_test,y_test,y_train), name="pca")
```

setelah dilakukan PCA selanjutnya mengecek akurasi dengan  menggunakan metode KNN

```python
import orchest
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
import warnings
warnings.filterwarnings("ignore")
data = orchest.get_inputs()

X_train,X_test,y_test,y_train = data['pca']

# print(pca.explained_variance_ratio_)
# print(pca.singular_values_)
knn = KNeighborsClassifier(n_neighbors = 15)
knn.fit(X_train,y_train)
hasil = knn.score(X_test, y_test)
orchest.output(hasil, name="hasilknn")
```
